```python
# Import Pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PeopleDataset.csv')
df.head()

# Uncomment the following lines to see more information about the DataFrame
# df.info()
# df.describe()
```

```python
# Output the first names and heights in a bar chart
plt.bar(df['FirstName'], df['Height_cm'])
plt.xlabel('First Name')
plt.ylabel('Height (cm)')
plt.title('Heights of Individuals')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
print("Total People:", len(df))
print("Average Age (Mean):", df['Age'].mean())
print("Minimum Age:", df['Age'].min())
print("Maximum Height:", df['Height_cm'].max())
print("Males:", df['Gender'].value_counts()['Male'])
print("Females:", df['Gender'].value_counts()['Female'])
print("Over 50s, owns a dog:", df['Age'].value_counts())[df['Age'] > 50 & df['HasDog'] == 'True'].sum()
```
