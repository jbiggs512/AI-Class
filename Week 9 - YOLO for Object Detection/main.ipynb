```python
import kagglehub
import shutil
import os

# Define dataset path
construction_dataset = "./data/construction-site-safety-image-dataset-roboflow"

# Download - doesn't support path parameter
download_path = kagglehub.dataset_download(
    "snehilsanyal/construction-site-safety-image-dataset-roboflow"
)

# Create data directory if it doesn't exist
os.makedirs("./data", exist_ok=True)

# Move the downloaded dataset folder into ./data
shutil.move(download_path, construction_dataset)

print("Downloaded to:", construction_dataset)
```
